 Real-Life Stock Market Example: Share Prices of a Tech Company
Imagine you're interested in the daily closing prices of a tech company's stock (like Apple or Tesla) over the past year (365 days).

üéØ What You Want to Know:
"What's the average stock price for the company?"

But instead of using all 365 days, you:

Take samples of 30 days at a time

Calculate the average closing price of those 30 days

Repeat this 100 times with different sets of 30 days

The collection of these 100 average prices forms a sampling distribution.

‚û§ Main Goal:
We want to estimate an average (like average click rate or average stock price) for a very large group (population), but we can't look at every single item in the group.

So we ask:

"If I take many small random samples and calculate their averages, what will those averages look like?"

That collection of averages is called the sampling distribution, and it helps us predict or understand the whole group even if we only looked at small parts of it.

What is the Central Limit Theorem (CLT)? (Simple Explanation)
The Central Limit Theorem (CLT) says:

If you take lots of random samples from a big group of things (like people, numbers, prices), and you calculate the average for each sample, those averages will:

‚úÖ Start to look like a bell-shaped curve (normal distribution),
‚úÖ Even if the original data wasn't bell-shaped,
‚úÖ And this happens when the sample size is big enough (usually 30 or more).

üîç The Problem:
Each company wants to know:

"What is our average customer satisfaction rating?"

"Can we predict how most people feel, even if we can't ask everyone?"

But asking every customer is impossible. So we:

Take random samples of 30 customers from each brand.

Find the average rating for each sample.

Repeat this many times (like 1000 samples).

Plot those average ratings.



What CLT Shows:
The individual ratings may be very spread out or uneven.

But the average ratings from all the samples (sampling distribution) will:

Form a normal curve (bell shape),

Center around the true average for each company,

Help us compare the brands fairly.

```python
import numpy as np
import matplotlib.pyplot as plt

# üéØ Step 1: Simulate the population - 365 days of stock prices
np.random.seed(42)
stock_prices = np.random.normal(loc=150, scale=15, size=365)  # Avg price: $150, SD: $15

# üéØ Step 2: Sampling 100 times, each time sample 30 days
sample_means = []

for _ in range(100):
    sample = np.random.choice(stock_prices, size=30)
    sample_mean = np.mean(sample)
    sample_means.append(sample_mean)

# üéØ Step 3: Plot the sampling distribution of average stock prices
plt.hist(sample_means, bins=10, edgecolor='black')
plt.title("Sampling Distribution of Stock Price (Sample Size = 30)")
plt.xlabel("Average Stock Price ($)")
plt.ylabel("Number of Samples")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Simulate customer ratings for each brand
np.random.seed(10)
ratings_apple = np.random.randint(3, 6, size=10000)    # Mostly good ratings
ratings_samsung = np.random.randint(2, 6, size=10000)  # Mixed ratings
ratings_ibm = np.random.randint(1, 5, size=10000)      # More lower ratings

# Step 2: Function to sample and get average ratings
def get_sample_means(data, num_samples=1000, sample_size=30):
    sample_means = []
    for _ in range(num_samples):
        sample = np.random.choice(data, size=sample_size)
        sample_means.append(np.mean(sample))
    return sample_means

# Step 3: Get sampling distributions
apple_means = get_sample_means(ratings_apple)
samsung_means = get_sample_means(ratings_samsung)
ibm_means = get_sample_means(ratings_ibm)

# Step 4: Plot
plt.figure(figsize=(12, 6))
plt.hist(apple_means, bins=20, alpha=0.6, label='Apple', edgecolor='black')
plt.hist(samsung_means, bins=20, alpha=0.6, label='Samsung', edgecolor='black')
plt.hist(ibm_means, bins=20, alpha=0.6, label='IBM', edgecolor='black')
plt.title("CLT: Sampling Distribution of Customer Ratings")
plt.xlabel("Average Rating from Samples")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
puppies = np.array([1,1,0,1,0,0,1,0,1,0,1,1,1,0,0,1,0,0,1,0,0,1])

p = puppies.mean()
print('Mean:', p)
print('Standard Deviation:', puppies.std())
print('Variance:', puppies.var())
np.random.choice(puppies, size = (1,5), replace = True)
np.random.choice(puppies, size = (1,5), replace = True).mean()

sample_props = []
for i in range(10000):
  sample = np.random.choice(puppies, 5, replace = True)
  sample_props.append(sample.mean())
sample_props = np.array(sample_props)

sm = sample_props.mean()
print(sample_props)
print('Mean:', sm)
print('Standard Deviation:', sample_props.std())
print('Variance:', sample_props.var())

sample_props_20 = []
for i in range(10000):
  sample20 = np.random.choice(puppies, 20, replace = True)
  sample_props_20.append(sample20.mean())
sample_props_20 = np.array(sample_props_20)

sm_20 = sample_props_20.mean()
print(sample_props_20)
print('Mean:', sm_20)
print('Standard Deviation:', sample_props_20.std())
print('Variance:', sample_props_20.var())
```

```python
num = [1,10,50,100]
means = []
for j in num:
  np.random.seed(1)
  x = [np.mean(np.random.randint(-40, 40, j)) for _i in range(1000)]
  means.append(x)
k = 0

fig, ax = plt.subplots(2, 2,figsize = (6,6))
for i in range(0,2):
  for j in range(0,2):
    ax[i, j].hist(means[k], bins = 10, density = True)
    ax[i, j].set_title(label = num[k])
    k += 1
plt.show()
```
